```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from gym.envs.toy_text.frozen_lake import FrozenLakeEnv
```

```python
def init(env):
    """
    env: OpenAI Gym Environment
    """
    Q = np.zeros((env.nS, env.nA))
    return Q
```

```python
def policy(env, state, Q, epsilon):
    if np.random.rand() < epsilon: # epsilon-greedy policy
        return env.action_space.sample()
    else:
        return np.argmax(Q[state])
```

```python
episodes = 5000
epsilon = 0.1
alpha = 0.4
discount = 0.99

env = FrozenLakeEnv()
Q = init(env)        
```

```python
# This is a model free algor so learn while moving
episode_rewards = []
for e in range(episodes):
    state = env.reset()
    rewards = []
    while True:
        action = policy(env, state, Q, epsilon)
        q_now = Q[state][action] # Get current Q value
        
        next_state, reward, done, _ = env.step(action)
        rewards.append(reward)
        q_next = max(Q[next_state]) # Get maximum Q value of next actions (off-policy)
        
        Q[state][action] += alpha * (reward + discount*q_next - q_now)
        if done: # Q values of the terminal state is always 0.
            break
        state = next_state
    episode_rewards.append(np.mean(rewards))
```

```python
print(Q)
plt.plot(episode_rewards)
```

```python
done = False
state = env.reset()
while not done:
    env.render()
    action = policy(env, state, Q, 0.0)
    state, _, done, _ = env.step(action)
env.render()
```
